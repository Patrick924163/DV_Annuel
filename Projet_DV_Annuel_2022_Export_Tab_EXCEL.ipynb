```python
## Definition de la fonction COMP_FAM_MP
def COMP_FAM_MP(STATUT,NB_BC,NB_BE):
    if STATUT=="MP" and NB_BC==0 and NB_BE==0:
        return "MP_SEUL"
    elif STATUT=="MP" and NB_BC>0 and NB_BE==0:
        return "MP+BC"
    elif STATUT=="MP" and NB_BC==0 and NB_BE>0:
        return "MP+BE"
    elif STATUT=="MP" and NB_BC>0 and NB_BE>0:
        return "MP+BC+BE"
    else:
        return ""
```

```python
## définition de la fonction MOTIF_DV
def MOTIF_DV(MOTIF):
    if MOTIF in ["DV cnt indiv concurrent non réf","DV cnt indiv concurrent réf CNP","DV cnt indiv concurrent réf Intériale",
                "DV cnt indiv concurrent réf non précisé"]:
        return "Concurrence"
    elif MOTIF=="Contrat groupe Oblig. personnel":
        return "Contrat obligatoire personnel (ANI)"
    elif MOTIF=="Contrat groupe conjoint":
        return "Contrat groupe conjoint"
    elif MOTIF=="DV Adhérent injoignable":
        return "Adhérent injoignable"
    elif MOTIF=="DV Refus d'explication":
        return "Refus d'explication"
    elif MOTIF=="DV Rétention non réalisée":
        return "Rétention non réalisée"
    else:
        return "Autres"
```

```python
## Définition de la fonction TYPE_COTISATION_bis
def TYPE_COTISATION_bis(TYPE_DERNIERE_COTISATION):
    if TYPE_DERNIERE_COTISATION.strip()=="plancher":
        return "1. Plancher"
    elif TYPE_DERNIERE_COTISATION.strip() in ["forfait","taux","normal"]:
        return "2. Normal"
    elif TYPE_DERNIERE_COTISATION.strip() =="plafond":
        return "3. Plafond"
    else:
        return TYPE_DERNIERE_COTISATION
```

```python
## Définition de la fonction TYPE_COTISATION_ter
def TYPE_COTISATION_ter(TOP_COT):
    if TOP_COT == "plancher":
        return "1. Plancher"
    elif TOP_COT in ["forfait","taux","normal"]:
        return "2. Normal"
    elif TOP_COT == "plafond":
        return "3. Plafond"
    else:
        return TOP_COT
```

```python
## Fonction générale pour le calcul des statistiques (cette fonction va retourner un générateur)

## Ci-dessous lexique des paramètres :
## df = Source du dataframe sur laquelle sera utilisée pour calculer la statistique
## group = Liste des variables pour la formation des groupes (= groupe by)
## factor = 'NOIDE' ou 'EFFECTIF' (si variable indiquée = NOIDE --> compter ; si variable indiquée = EFFECTIF --> Sommer)
## num_stat = numéro de la statistique que l'on veut sortir
## filter_on = filtrage positif (Ex: ("STATUT",["MP", "MPA", "MPJ"]) ==> STATUT in ("MP", "MPA", "MPJ") ) --> équivalent IN
## filter_off = filtrage négatif  (Ex: (STATUT,["SSCR"]) ==> STATUT NE "SSCR" ) --> équivalent NOT IN

## La condition entre les différents filtrages est la relation AND

def Stat(df,group,factor,num_stat,filter_on=[],filter_off=[]):
    
    if len(filter_on)>0:   # Paramètre 'filtre_on' est une liste donc si len(filtre_on)>0 --> signifie au moins un élément dans la liste
        
        for i in range(len(filter_on)):
            
            df=df[df[filter_on[i][0]].isin(filter_on[i][1])]   # Ici c'est condition IN. Ex : [0] =  "STATUT" --> 1er élément de la liste [i] et [1] = ["MP","MPA","MPJ"] --> 2ème élément de la liste [i]
            
    if len(filter_off)>0:
        
        for j in range(len(filter_off)):
            
             df=df[~df[filter_off[j][0]].isin(filter_off[j][1])]  # Ici c'est condition NOT IN. Ex : [0] =  "STATUT" --> 1er élément de la liste [j] et [1] = ["MP","MPA","MPJ"] --> 2ème élément de la liste [j]
                
    if factor=="NOIDE":
        
        df1=df.groupby(group,dropna=False).count()[factor].to_frame().sort_index().reset_index().rename(columns={factor:"Effectif_NOIDE"})
        
    else:
        
        df1=df.groupby(group,dropna=False).sum()[factor].to_frame().sort_index().reset_index().rename(columns={factor:"Effectif"})
        
    yield ('stat_'+str(num_stat),df1)  # yield pour optimiser la memoire RAM
```
